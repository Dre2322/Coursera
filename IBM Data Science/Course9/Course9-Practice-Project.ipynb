````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
import seaborn as sns
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
````

## Load the Titanic dataset using Seaborn


````python
titanic = sns.load_dataset('titanic')
titanic.head()
````

#### Features to drop
`deck` has a lot of missing values so we'll drop it. `age` has quite a few missing values as well. Although it could be, `embarked` and `embark_town` don't seem relevant so we'll drop them as well. It's unclear what `alive` refers to so we'll ignore it.
#### Target
`survived` is our target class variable.


````python
features = ['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'class', 'who', 'adult_male', 'alone']
target = 'survived'

X = titanic[features]
y = titanic[target]
````

### Exercise 2. Split the data into training and testing sets
Don't forget to consider imbalance in the target


````python
# ---------------------------------------------------------
# Split the dataset into training and testing sets
# ---------------------------------------------------------
# - 80% of data used for training
# - 20% of data reserved for testing
# - Stratification ensures the class distribution
#   of 'survived' remains consistent in both sets
# - random_state is set for reproducibility
# ---------------------------------------------------------

X_train, X_test, y_train, y_test = train_test_split(
    X,                      # Feature matrix
    y,                      # Target vector
    test_size=0.2,          # 20% test split
    stratify=y,             # Preserve class distribution
    random_state=42         # Ensure reproducibility
)

# ---------------------------------------------------------
# Verify that class proportions are consistent
# between training and testing sets
# ---------------------------------------------------------

y_train.value_counts(normalize=True)
y_test.value_counts(normalize=True)

````

### Define preprocessing transformers for numerical and categorical features
#### Automatically detect numerical and categorical columns and assign them to separate numeric and categorical features


````python
# ---------------------------------------------------------
# Automatically detect feature types based on data types
# ---------------------------------------------------------
# Numerical features:
#   - Continuous or discrete numeric values
# Categorical features:
#   - Object or category data types requiring encoding
# ---------------------------------------------------------

numerical_features = X_train.select_dtypes(
    include=['number']
).columns.tolist()

categorical_features = X_train.select_dtypes(
    include=['object', 'category']
).columns.tolist()
````

#### Define separate preprocessing pipelines for both feature types


````python
# ---------------------------------------------------------
# Numerical feature preprocessing pipeline
# ---------------------------------------------------------
# Steps:
# 1. Impute missing values using the median
#    (robust to outliers)
# 2. Standardize features to mean=0, std=1
# ---------------------------------------------------------

numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# ---------------------------------------------------------
# Categorical feature preprocessing pipeline
# ---------------------------------------------------------
# Steps:
# 1. Impute missing values using the most frequent category
# 2. One-hot encode categorical variables
#    - handle_unknown='ignore' prevents runtime errors
# ---------------------------------------------------------

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
````

#### Combine the transformers into a single column transformer
We'll use the sklearn "column transformer" estimator to separately transform the features, which will then concatenate the output as a single feature space, ready for input to a machine learning estimator.


````python
# ---------------------------------------------------------
# Combine numerical and categorical preprocessing
# into a single ColumnTransformer
# ---------------------------------------------------------
# Each transformer is applied only to its
# corresponding feature subset, then concatenated
# ---------------------------------------------------------

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)
````

### Create a model pipeline
Now let's complete the model pipeline by combining the preprocessing with a Random Forest classifier


````python
# ---------------------------------------------------------
# Create the full machine learning pipeline
# ---------------------------------------------------------
# The pipeline ensures:
# - Preprocessing occurs inside cross-validation
# - No data leakage between training and testing
# - Clean, reproducible modeling workflow
# ---------------------------------------------------------

pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])
````

### Define a parameter grid 
We'll use the grid in a cross validation search to optimize the model


````python
# ---------------------------------------------------------
# Define hyperparameters for GridSearchCV
# ---------------------------------------------------------
# Parameters are prefixed with 'classifier__'
# to reference the model inside the pipeline
# ---------------------------------------------------------

param_grid = {
    'classifier__n_estimators': [50, 100],
    'classifier__max_depth': [None, 10, 20],
    'classifier__min_samples_split': [2, 5]
}
````

### Perform grid search cross-validation and fit the best model to the training data


````python
# ---------------------------------------------------------
# Define stratified cross-validation strategy
# ---------------------------------------------------------
# StratifiedKFold preserves class distribution
# across all folds during cross-validation
# ---------------------------------------------------------

cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)
````

### Exercise 3. Train the pipeline model 


````python
# ---------------------------------------------------------
# Perform grid search with cross-validation
# ---------------------------------------------------------
# - estimator: full preprocessing + model pipeline
# - scoring: accuracy
# - verbose: display training progress
# ---------------------------------------------------------

model = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv=cv,
    scoring='accuracy',
    verbose=2
)

# Fit the grid search model to training data
model.fit(X_train, y_train)
````

### Exercise 4. Get the model predictions from the grid search estimator on the unseen data
Also print a classification report


````python
# ---------------------------------------------------------
# Generate predictions on the unseen test dataset
# ---------------------------------------------------------

y_pred = model.predict(X_test)

# ---------------------------------------------------------
# Display classification performance metrics
# ---------------------------------------------------------
# Includes precision, recall, f1-score, and accuracy
# ---------------------------------------------------------

print(classification_report(y_test, y_pred))
````

## Feature importances
Let's figure out how to get the feature importances of our overall model.  You'll need to know how to do this for your final project.  
First, to obtain the categorical feature importances, we have to work our way backward through the modelling pipeline to associate the feature importances with their one-hot encoded input features that were transformed from the original categorical features.

We don't need to trace back through the pipeline for the numerical features, because we didn't transfrom them into new ones in any way.  
Remember, we went from categorical features to one-hot encoded features, using the 'cat' column transformer.

Here's how you trace back through the trained model to access the one-hot encoded feature names:


````python
# ---------------------------------------------------------
# Retrieve one-hot encoded feature names
# ---------------------------------------------------------
# After OneHotEncoding, each categorical feature expands
# into multiple binary indicator columns.
# This pulls the transformed column names so we can align
# feature importances correctly.
# ---------------------------------------------------------

ohe_feature_names = (
    model.best_estimator_['preprocessor']
    .named_transformers_['cat']
    .named_steps['onehot']
    .get_feature_names_out(categorical_features)
)
````

Notice how the one-hot encoded features are named - for example, `sex` was split into two boolean features indicating whether the sex is male or female.

Great! Now let's get all of the feature importances and associate them with their transformed feature names.


````python
# ---------------------------------------------------------
# Extract feature importances from the trained classifier
# ---------------------------------------------------------
# RandomForestClassifier provides feature_importances_
# which reflects the average impurity decrease contributed
# by each feature across trees.
# ---------------------------------------------------------

feature_importances = model.best_estimator_['classifier'].feature_importances_

# Combine numeric features + one-hot encoded categorical features
feature_names = numerical_features + list(ohe_feature_names)

# Build a DataFrame for easy sorting/plotting
importance_df = (
    pd.DataFrame({
        'Feature': feature_names,
        'Importance': feature_importances
    })
    .sort_values(by='Importance', ascending=False)
)
````

### Display the feature importances in a bar plot
Define a feature importance DataFrame, then plot it 


````python
# ---------------------------------------------------------
# Plot feature importances
# ---------------------------------------------------------
# Note: plotting the top N features improves readability
# since one-hot encoding can create many columns.
# ---------------------------------------------------------

top_n = 15
top_features = importance_df.head(top_n)

plt.figure(figsize=(10, 6))
plt.barh(top_features['Feature'], top_features['Importance'])
plt.gca().invert_yaxis()
plt.title('Top Feature Importances (Random Forest)')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()

# ---------------------------------------------------------
# Report test set accuracy for the best estimator
# ---------------------------------------------------------

test_score = model.score(X_test, y_test)
print(f"\nTest set accuracy: {test_score:.2%}")

````

## Try another model
In practice you would want to try out different models and even revisit the data analysis to improve
your model performance. Maybe you can engineer new features or impute missing values to be able to use more data.

With Scikit-learn's powerful pipeline class, this is easy to do in a few steps.
Let's update the pipeline and the parameter grid so we can train a Logistic Regression model and compare the performance of the two models.


````python
# ---------------------------------------------------------
# Try another model: Logistic Regression
# ---------------------------------------------------------
# We will reuse the SAME preprocessor (numeric + categorical)
# but swap the classifier to Logistic Regression.
#
# Why create a NEW GridSearchCV object?
# - Avoids leftover state from the Random Forest grid search
# - Prevents confusion between param grids / best_estimator_
# - Easier to compare results side-by-side
# ---------------------------------------------------------

# 1) Build a new pipeline using the same preprocessing step
lr_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(random_state=42, max_iter=1000))
])

# 2) Define Logistic Regression hyperparameter grid
# Notes:
# - 'liblinear' supports l1 and l2 penalties
# - class_weight='balanced' can help with class imbalance
lr_param_grid = {
    'classifier__solver': ['liblinear'],
    'classifier__penalty': ['l1', 'l2'],
    'classifier__class_weight': [None, 'balanced']
}

# 3) Grid search with the same CV strategy
lr_model = GridSearchCV(
    estimator=lr_pipeline,
    param_grid=lr_param_grid,
    cv=cv,
    scoring='accuracy',
    verbose=2
)

# 4) Train
lr_model.fit(X_train, y_train)

# 5) Predict + Evaluate on test set
lr_y_pred = lr_model.predict(X_test)

print("Best LR Parameters:", lr_model.best_params_)
print("Best LR CV Score:", lr_model.best_score_)
print("\nLogistic Regression - Test Set Classification Report:")
print(classification_report(y_test, lr_y_pred))
````

### Exercise 7. Display the clasification report for the new model and compare the results to your previous model.


````python
# ---------------------------------------------------------
# Classification report for Logistic Regression
# ---------------------------------------------------------
# This report summarizes precision, recall, f1-score,
# and accuracy for each class on the test dataset.
# ---------------------------------------------------------

print(classification_report(y_test, lr_y_pred))

# ---------------------------------------------------------
# Compute test accuracies for both models
# ---------------------------------------------------------

# Random Forest (previous model)
rf_test_acc = model.score(X_test, y_test)

# Logistic Regression (new model)
lr_test_acc = lr_model.score(X_test, y_test)

print(f"Random Forest Test Accuracy: {rf_test_acc:.2%}")
print(f"Logistic Regression Test Accuracy: {lr_test_acc:.2%}")

````

### Extract the logistic regression feature coefficients and plot their magnitude in a bar chart.


````python
# ---------------------------------------------------------
# Extract Logistic Regression coefficients
# ---------------------------------------------------------
# coef_ shape: (1, n_features) for binary classification
# We extract the first (and only) row.
# ---------------------------------------------------------

coefficients = (
    lr_model
    .best_estimator_
    .named_steps['classifier']
    .coef_[0]
)

# ---------------------------------------------------------
# Retrieve transformed feature names
# ---------------------------------------------------------
# Numerical features remain unchanged
# Categorical features were expanded via OneHotEncoder
# ---------------------------------------------------------

numerical_feature_names = numerical_features

categorical_feature_names = (
    lr_model
    .best_estimator_
    .named_steps['preprocessor']
    .named_transformers_['cat']
    .named_steps['onehot']
    .get_feature_names_out(categorical_features)
)

# Combine all feature names in correct order
feature_names = numerical_feature_names + list(categorical_feature_names)
````

### Exercise 9. Plot the feature coefficient magnitudes in a bar chart
What's different about this chart than the feature importance chart for the Random Forest classifier?


````python
# ---------------------------------------------------------
# Create DataFrame of coefficients
# ---------------------------------------------------------
# - Coefficient sign indicates direction of effect
# - Absolute value indicates strength of influence
# ---------------------------------------------------------

coef_df = (
    pd.DataFrame({
        'Feature': feature_names,
        'Coefficient': coefficients
    })
    .assign(Magnitude=lambda df: df['Coefficient'].abs())
    .sort_values(by='Magnitude', ascending=False)
)

# ---------------------------------------------------------
# Plot top Logistic Regression feature magnitudes
# ---------------------------------------------------------
# Plotting top N improves readability since one-hot
# encoding can create many features.
# ---------------------------------------------------------

top_n = 15
top_coef_df = coef_df.head(top_n)

plt.figure(figsize=(10, 6))
plt.barh(top_coef_df['Feature'], top_coef_df['Magnitude'])
plt.gca().invert_yaxis()
plt.title('Top Logistic Regression Feature Coefficients (Magnitude)')
plt.xlabel('Absolute Coefficient Value')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
````

The Logistic Regression coefficient magnitude plot shows that gender-related features are the strongest predictors of survival, with who_man having the largest impact, indicating a significantly lower likelihood of survival for men. Passenger class (pclass) is the next most influential feature, highlighting the importance of socioeconomic status. Overall, the model reflects the historical “women and children first” pattern observed during the Titanic disaster and provides clear, interpretable insights into survival outcomes.
