# Module 2: Data Wrangling


### Question 2

Drop the columns <code>"id"</code>  and <code>"Unnamed: 0"</code> from axis 1 using the method <code>drop()</code>, then use the method <code>describe()</code> to obtain a statistical summary of the data. Make sure the <code>inplace</code> parameter is set to <code>True</code>. Take a screenshot of your code and output. You will need to submit the screenshot for the final project. 


```python
# -------------------------------------------------------------
# DROP COLUMNS "id" AND "Unnamed: 0"
# -------------------------------------------------------------
# We remove these columns because they are identifiers and do not
# contribute to predicting house price. inplace=True modifies df directly.

df.drop(["id", "Unnamed: 0"], axis=1, inplace=True)

# Display statistical summary after dropping the columns
df.describe()
```

We can see we have missing values for the columns <code> bedrooms</code>  and <code> bathrooms </code>


```python
# -------------------------------------------------------------
# CHECK FOR MISSING VALUES IN bedrooms AND bathrooms COLUMNS
# -------------------------------------------------------------

print("Number of NaN values in 'bedrooms'  :", df['bedrooms'].isnull().sum())
print("Number of NaN values in 'bathrooms' :", df['bathrooms'].isnull().sum())
```

We can replace the missing values of the column <code>'bedrooms'</code> with the mean of the column  <code>'bedrooms' </code> using the method <code>replace()</code>. Don't forget to set the <code>inplace</code> parameter to <code>True</code>


```python
# -------------------------------------------------------------
# FILL MISSING VALUES IN 'bedrooms' USING THE COLUMN MEAN
# -------------------------------------------------------------
# We compute the mean and replace NaN values with it.
# inplace=True ensures updates apply directly to df.

mean_bedrooms = df["bedrooms"].mean()
df["bedrooms"].replace(np.nan, mean_bedrooms, inplace=True)
```

We also replace the missing values of the column <code>'bathrooms'</code> with the mean of the column  <code>'bathrooms' </code> using the method <code>replace()</code>. Don't forget to set the <code> inplace </code>  parameter top <code> True </code>


```python
# -------------------------------------------------------------
# REPLACE MISSING VALUES IN 'bathrooms' WITH THE COLUMN MEAN
# -------------------------------------------------------------

# Calculate the mean value of the 'bathrooms' column
mean_bathrooms = df['bathrooms'].mean()

# Replace NaN values with the computed mean
df['bathrooms'].replace(np.nan, mean_bathrooms, inplace=True)
```

```python
# -------------------------------------------------------------
# VERIFY THAT NO MISSING VALUES REMAIN
# -------------------------------------------------------------

print("Number of NaN values for the column 'bedrooms'  :", df['bedrooms'].isnull().sum())
print("Number of NaN values for the column 'bathrooms' :", df['bathrooms'].isnull().sum())
```

# Module 5: Model Evaluation and Refinement


Import the necessary modules:


```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
print("done")
```

We will split the data into training and testing sets:


```python
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]    
X = df[features]
Y = df['price']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=1)


print("number of test samples:", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```

### Question 9

Create and fit a Ridge regression object using the training data, set the regularization parameter to 0.1, and calculate the R^2 using the test data. Take a screenshot of your code and the value of the R^2. You will need to submit it for the final project.


```python
from sklearn.linear_model import Ridge
```

```python
# -------------------------------------------------------------
# Create and fit a Ridge regression model with alpha = 0.1
# using the training data, then compute the R² on the test data.
# -------------------------------------------------------------

# Create Ridge regression object with regularization parameter alpha = 0.1
ridge_model = Ridge(alpha=0.1)

# Fit the model using the training data
ridge_model.fit(x_train, y_train)

# Calculate R² score on the test data
ridge_r2 = ridge_model.score(x_test, y_test)

ridge_r2
```

### Question 10

Perform a second order polynomial transform on both the training data and testing data. Create and fit a Ridge regression object using the training data, set the regularisation parameter to 0.1, and calculate the R^2 utilising the test data provided. Take a screenshot of your code and the R^2. You will need to submit it for the final project.


```python
# -------------------------------------------------------------
# Apply a degree-2 polynomial transform to the features, then
# fit a Ridge regression model (alpha = 0.1) and compute R² on
# the transformed test data.
# -------------------------------------------------------------

# Create the polynomial feature transformer (degree = 2)
poly = PolynomialFeatures(degree=2, include_bias=False)

# Transform training and test feature sets
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)

# Create Ridge regression model with same alpha as before
ridge_poly_model = Ridge(alpha=0.1)

# Fit the model on the polynomial-transformed training data
ridge_poly_model.fit(x_train_poly, y_train)

# Compute R² score on the polynomial-transformed test data
ridge_poly_r2 = ridge_poly_model.score(x_test_poly, y_test)

ridge_poly_r2
```
