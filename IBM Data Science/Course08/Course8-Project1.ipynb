### TASK 1.1: Develop a *Line chart* using the functionality of pandas to show how automobile sales fluctuate from year to year

````python
"""
TASK 1.1
-------------------------------------------------------------------------------
Create a line chart showing how Automobile Sales change from year to year.
We aggregate (sum) monthly automobile sales to obtain annual totals.

This visualization helps us understand:
 long-term automotive trends
 growth or decline over decades
 possible effects of recessions
 effects of key market events (2008 crisis, Covid-19, etc.)
-------------------------------------------------------------------------------
"""

# Convert Date column to actual datetime format so we can extract year
df['Date'] = pd.to_datetime(df['Date'])

# Extract numeric year value from the Date column
df['Year'] = df['Date'].dt.year

# Group monthly sales into annual totals
# We use sum() because each row is monthly data
yearly_sales = df.groupby('Year')['Automobile_Sales'].sum()

# Create a line plot
plt.figure(figsize=(12,6))   # wider plot

plt.plot(
    yearly_sales.index,          # X-axis = years
    yearly_sales.values,         # Y-axis = total vehicles sold
    marker='o',                  # add dots on each year
    linestyle='-',               # smooth line
    label="Annual Automobile Sales"
)

# Title + axis labels
plt.title("Historical Automobile Sales by Year", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Total Vehicles Sold")

# Add grid to improve readability
plt.grid(True, linestyle='--', alpha=0.6)

# Optional: add legend
plt.legend()

# Show the resulting visualization
plt.show()
````

````python
"""
Line Plot (Recession Highlighted)
--------------------------------------------
This visualization plots total annual automobile sales from 19802023.
Recession periods are highlighted using shaded vertical spans. In addition,
at least two recession years are annotated for visibility. All x-ticks
display individual years as requested.
"""

# ---- Convert datetime to year if needed ----
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year

# ---- Aggregate monthly sales into annual totals ----
yearly_sales = df.groupby('Year')['Automobile_Sales'].sum()

# ---- Manually define recession periods based on assignment description ----
recessions = [
    (1980, 1980),
    (1981, 1982),
    (1991, 1991),
    (2000, 2001),
    (2007, 2009),
    (2020, 2020) # Feb-Apr but rounded for yearly chart
]

# ---- Build plot ----
plt.figure(figsize=(16,6))
plt.plot(yearly_sales.index, yearly_sales.values, marker='o', label="Annual Sales")

# ---- Shade recession years ----
for start, end in recessions:
    plt.axvspan(start, end, color='red', alpha=0.15)

# ---- Annotate at least two recession years ----
plt.text(1980, yearly_sales.loc[1980], '1980 Recession', color='red')
plt.text(2008, yearly_sales.loc[2008], '2008 Crisis', color='red')

# ---- Formatting ----
plt.title("Automobile Sales during Recession", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Total Automobile Sales")
plt.grid(True, linestyle='--', alpha=0.5)

# ---- Display every year on x-axis ----
plt.xticks(yearly_sales.index, rotation=45)

plt.legend()
plt.tight_layout()
plt.show()
````

````python
"""
TASK 1.2
---------------------------------------------------------------------------
Goal:
Plot different line plots for each Vehicle_Type, and analyze whether the
sales trends differ between categories during recession periods.

Vehicle types include:
- Superminicar
- Smallfamilycar
- Mediumfamilycar
- Executivecar
- Sports

Method:
1) Extract annual sales by vehicle type.
2) Plot each type as a separate line.
3) Highlight recession years for comparison.
4) Visually analyze behavior during recessions.
---------------------------------------------------------------------------
"""

# aggregate monthly sales into annual sums by vehicle type
vehicle_year = df.groupby(["Year","Vehicle_Type"])["Automobile_Sales"].sum().reset_index()

# pivot year across columns for plotting
pivot_vehicle = vehicle_year.pivot(index="Year", columns="Vehicle_Type", values="Automobile_Sales")

# recession periods (same as previous task)
recessions = [
    (1980, 1980),
    (1981, 1982),
    (1991, 1991),
    (2000, 2001),
    (2007, 2009),
    (2020, 2020)
]

plt.figure(figsize=(16,6))

# plot each column as a separate line
for col in pivot_vehicle.columns:
    plt.plot(pivot_vehicle.index, pivot_vehicle[col], label=col)

# shade recession years
for start, end in recessions:
    plt.axvspan(start, end, color='red', alpha=0.15)

# formatting
plt.title("Automobile Sales by Vehicle Type During Recessions")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True, linestyle='--', alpha=0.5)

plt.xticks(pivot_vehicle.index, rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
````

### TASK 1.3: Use the functionality of **Seaborn Library** to create a visualization to compare the sales trend per vehicle type for a recession period with a non-recession period.


<details><summary>Click here for a hint</summary>

<p>
     To visualize the average number of vehicles sold during recession and non-recession periods, you can use a bar chart
        <br> You will need to group recession average Automobile_Sales and then plot it<br>
    Make use of sns.barplot(x=x,y=y, data = df)</p>
</details>


````python
"""
TASK 1.3
----------------------------------------------------------------------
Goal:
Compare the average automobile sales during recession vs non-recession
periods using a Seaborn bar chart.

Method:
1) Group automobile sales by the binary recession variable
2) Compute the mean sales in each category
3) Plot using seaborn barplot
----------------------------------------------------------------------
"""

# Calculate mean sales during recession and non-recession
new_df = df.groupby('Recession')['Automobile_Sales'].mean().reset_index()

# Create bar chart using seaborn
plt.figure(figsize=(8,5))
sns.barplot(
    x='Recession',
    y='Automobile_Sales',
    hue='Recession',
    data=new_df,
    palette='viridis'
)

# axis labels + title
plt.xlabel('Economic Condition')
plt.ylabel('Average Automobile Sales')
plt.title('Average Automobile Sales during Recession vs Non-Recession')

# custom category labels
plt.xticks(
    ticks=[0, 1],
    labels=['Non-Recession', 'Recession']
)

plt.legend().remove()
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.show()
````

### TASK 1.4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, you want to find more insights from the data to understand the reason. <br>Plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods? 
<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


````python
"""
TASK 1.4
---------------------------------------------------------------------
Goal:
Use matplotlib subplots to compare how GDP changes over time during
recession vs non-recession periods.

Steps:
1) Split the dataframe into recession and non-recession subsets
2) Aggregate monthly GDP into yearly values
3) Create two line plots using add_subplot()
---------------------------------------------------------------------
"""

# Split into recession and non-recession dataframes
df_recession = df[df['Recession'] == 1]
df_non_recession = df[df['Recession'] == 0]

# Aggregate GDP by year
recession_gdp = df_recession.groupby('Year')['GDP'].mean()
nonrecession_gdp = df_non_recession.groupby('Year')['GDP'].mean()

# Create figure
plt.figure(figsize=(16,5))

# ---- subplot 1: recession
ax1 = plt.subplot(1, 2, 1)
ax1.plot(recession_gdp.index, recession_gdp.values, marker='o', color='red')
ax1.set_title('GDP Trend during Recession Years')
ax1.set_xlabel('Year')
ax1.set_ylabel('GDP')
ax1.grid(True, linestyle='--', alpha=0.5)

# ---- subplot 2: non-recession
ax2 = plt.subplot(1, 2, 2)
ax2.plot(nonrecession_gdp.index, nonrecession_gdp.values, marker='o', color='green')
ax2.set_title('GDP Trend during Non-Recession Years')
ax2.set_xlabel('Year')
ax2.set_ylabel('GDP')
ax2.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
````

````python
city_to_state = {
    'New York'   : 'NY',
    'Los Angeles': 'CA',
    'Chicago'    : 'IL',
    'Houston'    : 'TX',
    'Phoenix'    : 'AZ',
    # TODO: add every city that appears in df['City']
    # missing ones that were actually states
    'Georgia'      : 'GA',
    'Illinois'     : 'IL',
    'California'   : 'CA'

}

# Create a new State column using the mapping
df['State'] = df['City'].map(city_to_state)
````

````python
"""
Choropleth map of automobile sales by U.S. state during a recession period.

Prerequisites:
    pip install plotly
    us-states.json must be in the same folder as this script/notebook.
"""

import json
import pandas as pd
import plotly.express as px

# --------------------------------------------------
# 1. Load your main data
# --------------------------------------------------
# If you already have df in memory, skip this read_csv line.
# df = pd.read_csv("your_automobile_sales_file.csv")

# --------------------------------------------------
# 2. Filter to recession period and aggregate by state
# --------------------------------------------------
# Keep only the rows where Recession == 1 (recession period)
recession_df = df[df["Recession"] == 1]

# Aggregate total (or average) sales per state during recession
state_sales = (
    recession_df
    .groupby("State")["Automobile_Sales"]
    .sum()  # or .mean() depending on what you want
    .reset_index()
)

# Optional: rename column for clarity
state_sales.rename(columns={"Automobile_Sales": "Total_Sales"}, inplace=True)

# --------------------------------------------------
# 3. Load the GeoJSON of US states
# --------------------------------------------------
with open("us-states.json", "r") as f:
    us_states_geojson = json.load(f)

# In this file:
#   - Feature "id" is the 2-letter state code (CA, TX, etc.)
#   - Feature "properties.name" is the full state name

# --------------------------------------------------
# 4. Build the choropleth map with Plotly Express
# --------------------------------------------------
fig = px.choropleth(
    state_sales,
    geojson=us_states_geojson,
    # Column in state_sales that holds the state code
    locations="State",
    # Key in the GeoJSON features that matches "State"
    featureidkey="id",          # use "properties.name" if you have full names
    color="Total_Sales",        # the value to color by
    color_continuous_scale="Blues",
    scope="usa",                # focus on USA
    labels={"Total_Sales": "Total Sales"},
    title="Automobile Sales by State During Recession"
)

# Make the map fit tightly around the data
fig.update_geos(fitbounds="locations", visible=False)

# --------------------------------------------------
# 5. Display the figure (works in VS Code/Jupyter)
# --------------------------------------------------
fig.show()
````
